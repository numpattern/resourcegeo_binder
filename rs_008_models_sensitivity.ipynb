**Grouped boxplots to check sensitivity in models**

```python
import resourcegeo as rs
df = rs.BaseData('assay_geo').data
df2 = df.loc[df['UNIT'].isin([f'{i}' for i in range(1,9)]  + ['OVB'])].copy()
df3 = df2.copy()
df3['CUpc'] = df3['CUpc'] +0.1
df4 = df2.copy()
df4['CUpc'] = df3['CUpc'] +0.2
dfs = [df2,df3,df4]
rs.boxplot_by_category_sensitivity(dfs,'CUpc','UNIT',figsize=(14,4),bulkstats=True,ylim=(-2,8),stats_ycoord=-1.8)
```
